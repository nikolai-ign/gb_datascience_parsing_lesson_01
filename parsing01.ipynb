```python
# 2. Работа будет состоять с недокументированным API. 
# Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search
# (а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. 
# Для каждой категории рассчитать среднюю минимальную стоимость заказа.
# (б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

# при загрузке страницы https://www.delivery-club.ru/search?vendorListQuery=пицца
# просматривая переданные данные по сети с фильтром по входжению "api"
# вижу интересный отправленный запрос
#https://api.delivery-club.ru/api1.2/vendors/search?latitude=55.7577374&longitude=37.6164793&query=пицца&cacheBreaker=1664008494
# ответом сервера на него является json большого размера, возможно это именно то что мне нужно.

url="https://api.delivery-club.ru/api1.2/vendors/search"
params={'latitude': '55.7577374',
        'longitude': '37.6164793',
        'query': 'пицца',
        'cacheBreaker': '1664008494'
       }
res = requests.get(url=url, params=params)

```

```python
print(res.status_code)
pprint.pprint(res.json())
```

```python
# формируем список словарей - название ресторана, стоимость доставки, минимальная сумма заказа, категория ресторан/магазин
# исходя из названия могу предположить что категория=1 - ресторан, категория=3 - магазин
vendors_list=[]
for vendor in res.json()['vendors']:
    dict1={'name':vendor['name'], 'delivery_price':vendor['delivery']['price']['value'], 'delivery_min_order':vendor['delivery']['minOrderPrice']['value'], 'category_id':vendor['categoryId'] }
    vendors_list.append(dict1)
    print(dict1)
#print(vendors_list)    
```

```python
delivery_free_rest=0 # количество ресторанов/магазинов с бесплатной доставкой
delivery_pay_rest=0 # количество ресторанов/магазинов с платной доставкой
delivery_free_shops=0 # количество магазинов с бесплатной доставкой
delivery_pay_shops=0 # количество магазинов с платной доставкой
delivery_free_sum_order=0 # бесплатная доставка - минимальная сумма заказа - нарастающий итог 
delivery_pay_sum_order=0 # платная доставка - минимальная сумма заказа - нарастающий итог
delivery_free_avg_sum_order=0 # бесплатная доставка - средняя минимальная сумма заказа
delivery_pay_avg_sum_order=0 # платная доставка - средняя минимальная сумма заказа


for vendor in vendors_list:
    if(vendor['delivery_price']==0):
        delivery_free_rest+=1
        delivery_free_sum_order+=vendor['delivery_min_order']
        if(vendor['category_id']==3):
            delivery_free_shops+=1
    else:
        delivery_pay_rest+=1
        delivery_pay_sum_order+=vendor['delivery_min_order']
        if(vendor['category_id']==3):
            delivery_pay_shops+=1

if(delivery_free_rest!=0):
    delivery_free_avg_sum_order=round(delivery_free_sum_order/delivery_free_rest,0)
if(delivery_pay_rest!=0):
    delivery_pay_avg_sum_order=round(delivery_pay_sum_order/delivery_pay_rest,0)
    
# выводим результат расчёта
print(f"Общее количество ресторанов/магазинов: {len(vendors_list)}")        
print("---")
print(f"Количество ресторанов/магазинов с бесплатной доставкой: {delivery_free_rest}")
print(f"Из них количество магазинов с бесплатной доставкой: {delivery_free_shops}")
print(f"Процент ресторанов/магазинов с бесплатной доставкой: {round((delivery_free_rest/len(vendors_list))*100,2)}")
print(f"Средняя минимальная сумма заказа из ресторанов/магазинов с бесплатной доставкой: {delivery_free_avg_sum_order}")
print("---")
print(f"Количество ресторанов/магазинов с платной доставкой: {delivery_pay_rest}")
print(f"Из них количество магазинов с платной доставкой: {delivery_pay_shops}")
print(f"Процент ресторанов/магазинов с бесплатной доставкой: {round((delivery_pay_rest/len(vendors_list))*100,2)}")
print(f"Средняя минимальная сумма заказа из ресторанов/магазинов с платной доставкой: {delivery_pay_avg_sum_order}")
   
```
